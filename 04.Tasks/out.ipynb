## Задача 1
### 1.1. Скачать данные по ссылке

````python
import pandas as pd
df = pd.read_csv("https://gbcdn.mrgcdn.ru/uploads/asset/4266730/attachment/08ec55854637add5247d22396d0f7456.csv")
df.info()
````

### 1.2 Изучите стоимости недвижимости

````python
df.describe()
````

````python
df["price"].min(), df["price"].mean().round(2), df["price"].max()
````

````python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,4))
plt.hist(df["price"])
plt.title('Распределение цены')
plt.xlabel('Цена млн. $')
plt.ylabel('Количество объектов')
plt.show();
````

### 1.3 Изучите распределение квадратуры жилой

````python
plt.figure(figsize=(10,5))
plt.hist(df["sqft_living"])
plt.title('Распределение жилой квадратуры')
plt.xlabel('Кв.метры')
plt.ylabel('Количество объектов')
plt.show();
````

### 1.4 Изучите распределение года постройки

````python
plt.figure(figsize=(10,5))
plt.hist(df["yr_built"])
plt.title('Распределение года постройки')
plt.xlabel('Года')
plt.ylabel('Количество объектов')
plt.show();
````
