Т.к. мы ранее по динамике продаж по месяцам, выяснили, что город не курортный, а цены в три раза у воды выше, то скорее всего площадей для строительства рядом с водой мало. Это может быть по двум причинам:
- Город расположен около маленького водоема, либо через него протекает всего одна река.

#### 3.5 Графики распределения цен на недвижимость с подвалом и без него

````python

df["Basement"] = df["sqft_basement"].apply(lambda x: "Есть подвал" if x>0 else "Нет подвала")
sns.displot(data = df, x ="price", hue="Basement", aspect=2, height=5, kind="kde")
plt.title('Распределение цены домов с подвалом и без')
plt.xlabel('Цена долларов')
plt.ylabel('Плотность распределения')
plt.ticklabel_format(style="plain", useOffset=False)
plt.xticks(rotation=30);
````

На графике показаны две кривые распределения цен с подвалом и без него. 
- Зависимости стоимости жилья от наличия подвала нет. Следовательно земля в городе не каменистая, т.к. этот факт отразился бы на стоимости строительства. 
- Домов с подвалами заметно больше чем домов без подвалов. В США чем севернее, тем более холодные зимы. Т.к. отсутствует центральное отопление, то в северных штатах в подвалах домов, принято располагать частные котельные.  Следовательно, город расположен в северном штате.

Нью-Йорк является одним из самых дорогих городов США с более дорогой недвижимостью, чем в "нашем" городе. В Нью-Йорке скалистая порода, способная без свай выдержать вес громадных многоэтажек, благодаря этому в нем и соседнем Манхеттене самое большое количество небоскребов в США.
